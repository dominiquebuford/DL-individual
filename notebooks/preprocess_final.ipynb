```python
import numpy as np
import pandas as pd
import os
```

```python
def calcLengthOfBody(row, excluded_columns):
  x_values = [val for columnName, val in row.items() if (columnName not in excluded_columns) and (columnName.find('_X')!=-1)]
  y_values = [val for columnName, val in row.items() if (columnName not in excluded_columns) and (columnName.find('_Y')!=-1)]
  len_X = max(x_values) - min(x_values)
  len_Y = max(y_values) - min(y_values)
  return len_X, len_Y
```

```python
def normalize(row):
    row_df = pd.DataFrame(row).transpose()
    excluded_columns = ['center_X','center_Y','ID', 'NAME', 'Scale', 'Activity', 'Category']
    lenBody_X, lenBody_Y = calcLengthOfBody(row, excluded_columns)
    for column in row_df.columns:
        if column not in excluded_columns:
            if (column.find('_X') != -1) and (row_df[column]!=-1).all():
                row_df[column] = (row_df[column] - row_df['center_X'])/lenBody_X
            if (column.find('_Y') != -1) and (row_df[column]!=-1).all():
                row_df[column] = (row_df[column] - row_df['center_Y'])/lenBody_Y

    return row_df.squeeze()
```

```python
df = pd.read_csv("../data/original_annotations.csv")
df.drop(columns = ['ID', 'Scale', 'Category'], inplace = True)
print(df.shape)
keptImages = os.listdir('../data/classification_images')
#filter to only include the images I picked
df = df[df['NAME'].isin(keptImages)]
print(df.shape)
df_extra = pd.read_csv("../data/df_extra.csv")
df_combined = pd.concat([df, df_extra], ignore_index=True)
print(df_combined.shape)
df_normalized = df_combined.apply(normalize, axis = 1)
```
